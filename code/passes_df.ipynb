```python
import pandas as pd
from pandas.io.json import json_normalize
import ast
from os import listdir
from os.path import isfile, join
```

```python
pd.set_option('display.max_columns', 100)
```

```python
files = [f for f in listdir('../data/') if isfile(join('../data/', f))]
```

```python
team = []
outcome = []
period = []
x0 = []
y0 = []
x1 = []
y1 = []
id_match = []
possession = []
counter = 0
for file in files:
    counter += 1
    match = pd.read_json('{}{}'.format('../data/',file))
    passes = match[match['pass'].notnull()]
    passes = passes[['team','location','pass','period','possession']]
    passes[['x0','y0']] = pd.DataFrame(passes['location'].tolist(), index= passes.index)
    df1 = (pd.concat({i: json_normalize(x) for i, x in passes.pop('pass').items()})
         .reset_index(level=1, drop=True)
         .join(passes)
         .reset_index(drop=True))
    df2 = (pd.concat({i: json_normalize(x) for i, x in df1.pop('team').items()})
         .reset_index(level=1, drop=True)
         .join(df1)
         .reset_index(drop=True))
    df2[['x1','y1']] = pd.DataFrame(df2['end_location'].tolist(), index= df2.index)
    df3 = df2[['name','outcome.name','x0','y0','x1','y1','period','possession']]
    df3 = df3.fillna('Complete')
    df3['possession'] = df3['possession'].astype('str')
    df3['possession'] = df3['possession']+'-'+df3['name']
    df3['id_match'] = counter
    
    possession.append(df3['possession'])
    team.append(df3['name'])
    period.append(df3['period'])
    id_match.append(df3['id_match'])
    outcome.append(df3['outcome.name'])
    x0.append(df3['x0'])
    y0.append(df3['y0'])
    x1.append(df3['x1'])
    y1.append(df3['y1'])
    
possession = [item for sublist in possession for item in sublist]
period = [item for sublist in period for item in sublist]
team = [item for sublist in team for item in sublist]
outcome = [item for sublist in outcome for item in sublist]
x0 = [item for sublist in x0 for item in sublist]
y0 = [item for sublist in y0 for item in sublist]
x1 = [item for sublist in x1 for item in sublist]
y1 = [item for sublist in y1 for item in sublist]
id_match = [item for sublist in id_match for item in sublist]

passes = pd.DataFrame({'team':team,'period':period,'outcome':outcome,'x0':x0,'y0':y0,'x1':x1,'y1':y1,'possession':possession,'id_match':id_match})    
```

```python
passes.to_csv('../df/passes.csv')
```
