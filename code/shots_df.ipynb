```python
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
import ast
from os import listdir
from os.path import isfile, join
```

```python
files = [f for f in listdir('../data/') if isfile(join('../data/', f))]
```

```python
len(files)
```

```python
team = []
period = []
id_match = []
possession = []
xg = []
counter = 0
for file in files:
    counter += 1
    match = pd.read_json('{}{}'.format('../data/',file))
    shots = match[match['shot'].notnull()]
    shots = shots[['team','period','possession','shot']]
    df1 = (pd.concat({i: json_normalize(x) for i, x in shots.pop('shot').items()})
         .reset_index(level=1, drop=True)
         .join(shots)
         .reset_index(drop=True))
    df2 = (pd.concat({i: json_normalize(x) for i, x in df1.pop('team').items()})
         .reset_index(level=1, drop=True)
         .join(df1)
         .reset_index(drop=True))

    df3 = df2[['name','statsbomb_xg','period','possession']]
    df3['id_match'] = counter
    df3['possession'] = df3['possession'].astype('str')
    df3['possession'] = df3['possession']+'-'+df3['name']   
    
    team.append(df3['name'])
    period.append(df3['period'])
    id_match.append(df3['id_match'])
    possession.append(df3['possession'])
    xg.append(df3['statsbomb_xg'])
        
period = [item for sublist in period for item in sublist]
team = [item for sublist in team for item in sublist]
id_match = [item for sublist in id_match for item in sublist]
possession = [item for sublist in possession for item in sublist]
xg = [item for sublist in xg for item in sublist]

shots = pd.DataFrame({'team':team,'period':period,'id_match':id_match,'possession':possession,'xg':xg})    
```

```python
shots.to_csv('../df/shots.csv')
```
